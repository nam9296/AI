### 1-3. 다음 boxplot을 그리시오.

```python
(ggplot(birthwt , aes("race", "bwt")) + geom_boxplot())
```

```python
(ggplot(birthwt , aes("factor(race)", "bwt")) + geom_boxplot())
```

```python
(ggplot(birthwt , aes("factor(race)", "bwt", fill='race')) + geom_boxplot())
```

```python
birthwt['race'] = birthwt['race'].astype('object')
```

```python
(ggplot(birthwt , aes("race", "bwt", fill='race')) + geom_boxplot())
```

### 2-1 diamonds 데이터를 이용해 다음 bar_plot을 그리시오.

```python
diamonds.head()
```

```python
(ggplot(diamonds, aes('clarity', fill='cut')) +
    geom_bar(position='dodge') +coord_flip())
```

### 2-2 diamonds 데이터를 이용해 다음 비율을 나타내는 bar_plot을 그리시오.

```python
(ggplot(diamonds, aes(x = "cut", fill = "clarity")) +
    geom_bar(position = "fill"))
```

### 2-3 diamonds 데이터를 이용해 다음 point_plot을 그리시오.

```python
diamonds.head()
```

```python
p = ggplot(diamonds, aes('carat', 'price',color="color")) +geom_point(alpha = 0.3)
p
```

```python
p + facet_wrap('clarity')
```
